```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('prosperLoanData.csv')
```

```python
# high-level overview of data shape and composition
print(df.shape)
```

```python
df.head(10)
```

```python
# Create a new dataframe with features of interset
col =['Term','BorrowerAPR', 'BorrowerRate', 'ProsperRating (Alpha)', 'EmploymentStatus', 'StatedMonthlyIncome', 'LoanOriginalAmount', 
       'LoanOriginationDate','MonthlyLoanPayment']
df_n = df[col]
```

```python
df_n.head()
```

```python
df_n.info()
```

### Assessing  and Cleaning

```python
#Checking EmploymentStatus values
print(df_n['EmploymentStatus'].unique())
```

```python
# Fill nan values with 'Not available'
df_n['EmploymentStatus'].fillna(value='Not available', inplace=True)
```

```python
#Check
df_n.info()
```

- 25 missing values from BorrowerAPR.
- We are going to calculate the median differnce betrwee BorrowerAPR and BorrowerRate
- Add the the median to BorrowerRate to get the missing BorrwerAPR

```python
diff = df_n["BorrowerAPR"] - df_n["BorrowerRate"]
diff.median()
```

```python
df_n["BorrowerAPR"].fillna(df_n["BorrowerRate"] + diff.median(), inplace=True)

#Check 
df_n["BorrowerAPR"].isnull().sum()
```

```python
#Rename a ProsperRating column 
df_n.rename(columns={'ProsperRating (Alpha)': 'ProsperRating'}, inplace=True)
```

```python
#Check
df_n.info()
```

```python
#Dropping nan records 
df_n.dropna(inplace=True)
```

```python
df_n.info()
```

```python
#Convert LoanOriginationDate column to datetime type
df_n['LoanOriginationDate']= pd.to_datetime(df_n['LoanOriginationDate'])
```

```python
df_n.info()
```

## Univariate Exploration

> I'll start by looking at the distribution of the main variable of interest: APR

```python
# start with a standard-scaled plot
binsize = 0.025
bins = np.arange(0, df_n['BorrowerAPR'].max()+binsize, 0.01)

plt.figure(figsize=[8, 5])
plt.hist(data = df_n, x = 'BorrowerAPR', bins = bins)
plt.xlabel('APR')
plt.show()
```

The distribution of APR looks multimodal. A small peak centered around 0.1, a large peak around at 0.2. There is also a small peak centered 0.3. Additionally, there is a very shape peak between 0.35 and 0.36.

Next up, the first predictor variable of interest: LoanOriginalAmount

```python
bins = np.arange(1000, df_n.LoanOriginalAmount.max()+500, 1000)
plt.figure(figsize=[8, 5])
plt.hist(data = df_n, x = 'LoanOriginalAmount', bins = bins);
plt.xlabel('Original loan amount ($)');
```

There are spikes in frequency at 5k, 10k, 20k and 25k. 

I'll now move on to the other variable in the dataset: stated monthly income

```python
# Distribution of stated monthly income
bins = np.arange(0, 50000, 300)
plt.hist(data = df_n, x = 'StatedMonthlyIncome', bins=bins);
plt.xlabel('Monthly Income ($)');
```

The distribution of the monthly income is right skewed, with most of monthly income less than 10K.

Next up: MonthlyLoanPayment

```python
# Distribution of stated monthly Loan payment
bins = np.arange(0, df_n.MonthlyLoanPayment.max()+200, 50)
plt.hist(data = df_n, x = 'MonthlyLoanPayment', bins=bins);
plt.xlabel('Monthly Loan Payment ($)');
```

Monthly Loan payment distrubtion looks rightly skewed. Most of the payments are less then 500

Next up we will look at ProsperRating.

AA is the lowest risk down to HR which actually stands for high risk. Most accounts have average Prosper Rating.

Next up: Term

The length of most of the loans are 36 months.

Next: EmploymentStatus

Most of borrowers are employed

## Bivariate Exploration

> To start off with, I want to look at the pairwise correlations present between features in the data.

```python
numeric_vars  = ['LoanOriginalAmount', 'BorrowerAPR', 'StatedMonthlyIncome', 'MonthlyLoanPayment']
categoric_vars  = ['Term', 'ProsperRating', 'EmploymentStatus']
```

```python
# correlation plot
plt.figure(figsize = [8, 5])
sb.heatmap(df_n[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)
plt.show()
```

```python
# plot matrix: sample 1500 loans so that plots are clearer and
# they render faster
samples = np.random.choice(df_n.shape[0], 1500, replace = False)
df_n_sample = df_n.loc[samples,:]

g = sb.PairGrid(data = df_n_sample, vars = numeric_vars)
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter, alpha=0.2);
```

We found that there's correlation between borrower APR and loan amount. The correlation coefficient is -0.426. We also found that there's a strong positive relation between the loan amount and the state monthly income which expected since borrowers with high monthly income could ask for large loan amount. We also noticed there is a strong correlation between loan payment and loan amount.


Let's move on to looking at how borrower APR, stated monthly income and loan original amount correlate with the categorical variables.

```python
# plot matrix of numeric features against categorical features.

samples = np.random.choice(df_n.shape[0], 2000, replace = False)
df_n_samp = df_n.loc[samples,:]

def boxgrid(x, y, **kwargs):
    """ Quick hack for creating box plots with seaborn's PairGrid. """
    default_color = sb.color_palette()[0]
    sb.boxplot(x, y, color = default_color)

plt.figure(figsize = [8, 10])
g = sb.PairGrid(data = df_n_samp, y_vars =  ['BorrowerAPR', 'StatedMonthlyIncome', 'LoanOriginalAmount', 'MonthlyLoanPayment'], x_vars = categoric_vars,
                height = 3, aspect = 1.5)
g.map(boxgrid)
plt.xticks(rotation = 30);
plt.show();
```

It appears that the loan amount increases with the increase of term. The borrower APR decreases with high Prosper Rating.
We noticed borrowers with high Rating have large loan amount. Not employed and Retired have high borrower APR.


Finally, let's look at relationships between the three categorical features.

```python
# since there's only three subplots to create, using the full data should be fine.
plt.figure(figsize = [10, 10])

# subplot 1: ProsperRating vs term
plt.subplot(3, 1, 1)
sb.countplot(data = df_n, x = 'ProsperRating', hue = 'Term', palette = 'Blues')

# subplot 2: EmploymentStatus vs. Term
ax = plt.subplot(3, 1, 2)
sb.countplot(data = df_n, x = 'EmploymentStatus', hue = 'Term', palette = 'Blues')
ax.legend(ncol = 2) # re-arrange legend to reduce overlapping

# subplot 3: ProsperRating vs. EmploymentStatus, use different color palette
ax = plt.subplot(3, 1, 3)
sb.countplot(data = df_n, x = 'EmploymentStatus', hue = 'ProsperRating', palette = 'Greens')
ax.legend(loc = 1, ncol = 2) # re-arrange legend to remove overlapping

plt.show()
```

High Prosper Ratings (AA-A) loans have 36 months duration. There are more 60 month loans on B and C ratings. There is only 36 months loans for HR rating borrowers.
Employment Status feature doesn't provide us with any informatioin, since there is only few data about non employed people.

## Multivariate Exploration

> The main thing I want to explore in this part of the analysis is how the categorical variable (Prosper rating and Term) play into the relationship between borrower APR and loan original amount.

```python
## create faceted plot on levels of the Term variable
g=sb.FacetGrid(data=df_n, aspect=1, height=4, col='Term', col_wrap=4)
g.map(plt.scatter, 'LoanOriginalAmount', 'BorrowerAPR');
g.add_legend();
```

From the above plot, we can see that Term doesn't have an impact on loan amount and borrower APR

```python
# Prosper rating effect on relationship of APR and loan amount
g=sb.FacetGrid(data=df_n, aspect=1, height=5, col='ProsperRating', col_wrap=4)
g.map(plt.scatter, 'LoanOriginalAmount', 'BorrowerAPR');
g.add_legend();
```

With high Prosper Rating, the loan amount increases. Also the borrower APR decreases with high rating.
